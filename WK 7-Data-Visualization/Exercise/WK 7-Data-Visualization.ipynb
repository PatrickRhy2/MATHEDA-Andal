```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"..\raw\chip-dataset-cleaned.csv",
                 delimiter=",")
df.info()
```

```python
df = df.drop(df.columns[[10,11,12]],axis=1)
df.head()
```

# N Chips Manufactured

```python
df_w = df.copy()
```

```python
chip_count = df_w['Vendor'].value_counts()
chip_count
```

```python
type(chip_count)
```

```python
chart = chip_count.plot(kind='bar',
                color = ['#080838','#111154','#191970','#292C7B','#393F86'],
                figsize=(10,5))
chart.bar_label(chart.containers[0], label_type='edge')
plt.title('Chips Manufactured (CPU+GPU)')
plt.xlabel('')
chart.set_xticklabels(chip_count.index, rotation=0) 
plt.ylabel('N Chips')
```

# N CPU Manufactured

```python
df_cpu = df_w[df_w['Type']=='CPU']
df_cpu
```

```python
cpu_count = df_cpu['Vendor'].value_counts()
cpu_count
```

```python
chart = cpu_count.plot(kind='bar',
                       color = ['#080838'],
                       figsize=(6,3))
plt.title('CPU Manufactured')
plt.ylabel('N Count')
plt.xlabel('')
plt.show()
```

# Lowest and Highest TDP (Thermal Design Power)

```python
df_tdp = df_w[['Product','TDP']]
df_tdp
```

```python
type(df_tdp)
```

```python
df_tdp_asc = df_tdp.copy()
df_tdp_asc = df_tdp_asc.sort_values(by='TDP', ascending=True)
df_tdp_asc
```

```python
df_tdp_asc = df_tdp_asc.head(11)
df_tdp_asc
```

```python
plt.figure(figsize=(10,5))
plt.bar(df_tdp_asc['Product'],df_tdp_asc['TDP'],
        color = '#080838')
plt.title('Products with the Lowest TDP')
plt.ylabel('Watts')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
df_tdp_desc = df_tdp.copy()
df_tdp_desc = df_tdp_desc.sort_values(by='TDP', ascending=False)
df_tdp_desc
```

```python
df_tdp_desc = df_tdp_desc.head(11)
df_tdp_desc
```

```python
plt.figure(figsize=(10,5))
plt.bar(df_tdp_desc['Product'],df_tdp_desc['TDP'],
        color = '#080838')
plt.title('Products with the Highest TDP')
plt.ylabel('Watts')
plt.xticks(rotation=45, ha='right')
plt.show()
```

# GPU Process Size Trend

```python
df_ps = df_w[['Type','Date','Process_Size']].where(df_w['Type']=='GPU').dropna()
df_ps
```

```python
df_ps['Date'] = pd.to_datetime(df_ps['Date'])
df_ps['Year'] = df_ps['Date'].dt.year
df_ps 
```

```python
df_ps = df_ps.groupby('Year')['Process_Size'].mean().reset_index()
df_ps
```

```python
plt.figure(figsize=(10,5),)
plt.plot(df_ps['Year'],df_ps['Process_Size'],
         color = '#080838')
plt.title('GPU Process Size Trend')
plt.ylabel('nm')
```
